# Test with IMDB Dataset

```python
from datasets import load_dataset
# Load the IMDB dataset
dataset = load_dataset("imdb")
```

```python
subset = dataset['test'].shuffle(seed=42).select(range(100))

# Perform sentiment analysis on the subset
sentiments = [analyze_sentiment(review['text'],sentiment_pipeline) for review in subset]

# Print the original texts and their predicted sentiments
for review, sentiment in zip(subset['text'], sentiments):
    print(f"Review: {review[:200]}...")  # Print the first 200 characters of the review
    print(f"Predicted Sentiment: {sentiment}\n")
```
